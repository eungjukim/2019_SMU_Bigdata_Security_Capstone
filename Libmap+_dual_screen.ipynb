```python
import warnings
import json
import numpy as np
from folium.plugins import MarkerCluster, Search, MousePosition, Draw, MeasureControl
from folium import FeatureGroup, LayerControl, plugins

warnings.simplefilter(action = "ignore", category = FutureWarning)
 
geo_path = 'skorea-provinces-2018-geo.json'
geo_json = json.load(open(geo_path, encoding='utf-8'))

def lib_spot():
    #지도 생성
    lib_map = folium.plugins.DualMap(location=[36, 128], tiles = 'cartodbdark_matter', zoom_start = 7, control_scale=True)
    
    #마커 클러스터 생성
    marker_cluster = MarkerCluster(name = "도서관 Cluster").add_to(lib_map.m1)
    
    #feature group 생성
    feature_group = FeatureGroup(name='도서관 Marker')
    feature_group2 = FeatureGroup(name='도서관 상호대차')
    
    #참여도서관 수 단계구분도
    lib_map.m1.choropleth(geo_data = geo_json,
                       data = do_lib['참여도서관_수'],
                       columns = [do_lib.index, do_lib['참여도서관_수']],
                       key_on = 'feature.properties.name_eng',
                       fill_color = 'Greens',
                       fill_opacity = 0.7,
                       line_opacity = 0.2,
                       legend_name = '참여도서관 수',
                       highlight = True,
                       name = '도서관 수 단계구분도'
                      )
    #도서관 대비 인구수 단계구분도
    lib_map.m1.choropleth(geo_data = geo_json,
                       data = do_lib['도서관_대비_인구수'],
                       columns = [do_lib.index, do_lib['도서관_대비_인구수']],
                       key_on = 'feature.properties.name_eng',
                       fill_color = 'Blues',
                       fill_opacity = 0.7,
                       line_opacity = 0.2,
                       legend_name = '도서관 대비 인구수',
                       highlight = True,
                       name = '도서관 대비 인구수 단계구분도'
                      )
    
    #지도 위에 도서관 위치 표시
    length = x.shape[0]
    
    for i in range(0, length, 1):
        #마커 표시를 위한 데이터 생성
        lat = x.loc[i, "위도"]
        lon = x.loc[i, "경도"]
        lib_name = str(x.loc[i, "도서관명"])
        lib_add = str(x.loc[i, "주소"])
        lib_url = str("<a href=\"{}\">{}<a>").format(x.loc[i, "홈페이지"], x.loc[i, "홈페이지"])
        lib_call = str(x.loc[i, "전화번호"])
        popup_msg = ("<b>도서관명:<b> {name}<br>""<b>주소:<b> {address}</a> <br>""<b>URL:<b> {url}</a> <br>""<b>전화번호:<b> {call} <br>").format(name = lib_name, address = lib_add, url = lib_url, call = lib_call)
        
        #마커 표시   
        folium.CircleMarker([lat, lon], radius = 5, popup = popup_msg, fill = True, fill_color = "purple", color = "purple", fill_opacity = 0.5, tooltip = x.loc[i, "도서관명"]).add_to(feature_group)
        
        folium.CircleMarker([lat, lon], radius = 5, popup = popup_msg, fill = True, fill_color = "blue", color = "blue", tooltip = x.loc[i, "도서관명"]).add_to(marker_cluster)
       
    p_array = []

    #기준도서관 그룹화(2차원 배열)
    for i in range(0, int(max(mutual_loan['그룹번호'])), 1):
        inner_array = []

        for j in range(0, mutual_loan.shape[0], 1):

            if(mutual_loan.loc[j, '그룹번호']==i+1):
                p = [mutual_loan.loc[j, "위도"], mutual_loan.loc[j, "경도"]]
                inner_array.append(p)
        p_array.append(inner_array)

    for i in range(0, len(p_array), 1):
        for j in range(0, len(p_array[i]), 1):
            #folium.plugins.AntPath(
                #locations = [p_array[i][0], p_array[i][j]],
                #reverse = 'True',
                #dash_array = [20, 30],
                #color = 'skyblue'
                #).add_to(feature_group2)
            folium.PolyLine([p_array[i][0], p_array[i][j]], interactive = True, bubblingMouseEvents = True, color = 'skyblue', weight = 3).add_to(feature_group2)
    
    
    #IP를 통해 이용자가 접속한 위치 표시
    user_msg = "Your current location."
    user_location = folium.FeatureGroup(name='당신의 현재 위치(IP 기반)').add_to(lib_map)
    folium.Marker([user_lat, user_lon], popup = user_msg, icon = folium.Icon(color="red", icon="info-sign")).add_to(user_location)
    
    #feature group 추가
    feature_group.add_to(lib_map.m2)
    feature_group2.add_to(lib_map.m2)
    
    #미니맵 추가
    minimap = plugins.MiniMap()
    lib_map.add_child(minimap)
    
    #tile 종류 추가
    folium.TileLayer('openstreetmap').add_to(lib_map.m1)
    folium.TileLayer('cartodbpositron').add_to(lib_map.m1)
    folium.TileLayer('cartodbdark_matter').add_to(lib_map.m2)
    
    #구글맵 추가
    folium.raster_layers.TileLayer(
        tiles='http://{s}.google.com/vt/lyrs=s&x={x}&y={y}&z={z}',
        attr='google',
        name='google maps',
        max_zoom=20,
        subdomains=['mt0', 'mt1', 'mt2', 'mt3'],
        overlay=False,
        control=True,
    ).add_to(lib_map)

    folium.raster_layers.TileLayer(
        tiles='http://{s}.google.com/vt/lyrs=m&x={x}&y={y}&z={z}',
        attr='google',
        name='google street view',
        max_zoom=20,
        subdomains=['mt0', 'mt1', 'mt2', 'mt3'],
        overlay=False,
        control=True,
    ).add_to(lib_map)
    
    #tile layer 추가
    LayerControl().add_to(lib_map)

    #mouse position 추가
    formatter = "function(num) {return L.Util.formatNum(num, 3) + ' º ';};"
    MousePosition(position='bottomleft',
                  separator=' | ',
                  empty_string='NaN',
                  prefix = '좌표',
                  lat_formatter=formatter,
                  lng_formatter=formatter).add_to(lib_map)
    
    #full screen 추가
    plugins.Fullscreen(
        position='topleft',
        title='전체화면으로 표시',
        title_cancel='돌아가기',
        force_separate_button=True).add_to(lib_map)
    
    #그림 그리기 추가
    #draw = Draw()
    #draw.add_to(lib_map)
    
    #measure control 추가
    measure = MeasureControl(position = 'topleft')
    measure.add_to(lib_map)
    
    
    #HTML 파일로 저장
    lib_map.save("Libmap+_dual_screen.html")

lib_spot()
```
